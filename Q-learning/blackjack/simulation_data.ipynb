```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
import seaborn as sns
sns.set_style("ticks")
sns.set_context("notebook")
```

```python
df_rewards_mean_nr_ce = pd.read_csv("./data/df_rewards_mean_nr_ce.csv", index_col=0) #normal reward, constant epsilon
df_rewards_mean_nr_ed = pd.read_csv("./data/df_rewards_mean_nr_ed.csv", index_col=0) #normal reward, constant epsilon

df_rewards_mean_cr_ce = pd.read_csv("./data/df_rewards_mean_cr_ce.csv", index_col=0) #custom reward, constant epsilon
df_rewards_mean_cr_ed = pd.read_csv("./data/df_rewards_mean_cr_ed.csv", index_col=0) #custom reward, epsilon decay

df_succes_rate_nr_ce = pd.read_csv("./data/df_succes_rate_nr_ce.csv", index_col=0)
df_succes_rate_nr_ed = pd.read_csv("./data/df_succes_rate_nr_ed.csv", index_col=0)

df_succes_rate_cr_ce = pd.read_csv("./data/df_succes_rate_cr_ce.csv", index_col=0)
df_succes_rate_cr_ed = pd.read_csv("./data/df_succes_rate_cr_ed.csv", index_col=0)
```

```python
df_rewards_mean_nr_ed.head()
```

```python
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2, 2, figsize=(10,6), layout='constrained', sharex=True)
plt.rcParams.update({"text.usetex": True,"font.family": "Computer Modern Serif"})

ax1.plot(df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "lognn"]["Episodio"],
         df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "lognn"]["35"], label=r"$\epsilon = 0.30$")
ax1.plot(df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "lognn"]["Episodio"],
         df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "lognn"]["50"], label=r"$\epsilon = 0.50$")
ax1.plot(df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "lognn"]["Episodio"],
         df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "lognn"]["65"], label=r"$\epsilon = 0.65$")
ax1.legend(loc='best')
ax1.set_title(r"$\alpha_{n} = \frac{\log(n)}{n}$")
ax1.xaxis.set_major_locator(ticker.FixedLocator([100, 100000, 300000, 500000]))

ax2.plot(df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "AB"]["Episodio"],
         df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "AB"]["35"], label=r"$\varepsilon = 0.30$")
ax2.plot(df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "AB"]["Episodio"],
         df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "AB"]["50"], label=r"$\varepsilon = 0.50$")
ax2.plot(df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "AB"]["Episodio"],
         df_succes_rate_nr_ce.loc[df_succes_rate_nr_ce["Función paso"] == "AB"]["65"], label=r"$\varepsilon = 0.65$")
ax2.legend(loc='best')
ax2.set_title(r"$\alpha_{n} = \frac{A}{B+n}$, $A=5$, $B=10$")
ax2.xaxis.set_major_locator(ticker.FixedLocator([100, 100000, 300000, 500000]))

ax3.plot(df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "lognn"]["Episodio"],
         df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "lognn"]["ed"], label=r"$\varepsilon_{n} = 1 - \frac{1}{n}$")
ax3.plot(df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "lognn"]["Episodio"],
         df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "lognn"]["ed2"], label=r"$\varepsilon_{n} = \frac{1}{n}$")
ax3.plot(df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "lognn"]["Episodio"],
         df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "lognn"]["ed3"], label=r"$\varepsilon_{n} = \frac{1}{1+n}$")
ax3.legend(loc='best')
ax3.set_title(r"$\alpha_{n} = \frac{\log(n)}{n}$")
ax3.xaxis.set_major_locator(ticker.FixedLocator([100, 100000, 300000, 500000]))

ax4.plot(df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "AB"].iloc[0::2]["Episodio"],
         df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "AB"].iloc[0::2]["ed"], label=r"$\varepsilon_{n} = 1 - \frac{1}{n}$")
ax4.plot(df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "AB"].iloc[0::2]["Episodio"],
         df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "AB"].iloc[0::2]["ed2"], label=r"$\varepsilon_{n} = \frac{1}{n}$")
ax4.plot(df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "AB"].iloc[0::2]["Episodio"],
         df_succes_rate_nr_ed.loc[df_succes_rate_nr_ed["Función paso"] == "AB"].iloc[0::2]["ed3"], label=r"$\varepsilon_{n} = \frac{1}{1+n}$")
ax4.legend(loc='best')
ax4.set_title(r"$\alpha_{n} = \frac{A}{B+n}$, $A=5$, $B=10$")
ax4.xaxis.set_major_locator(ticker.FixedLocator([100, 100000, 300000, 500000]))

fig.supylabel("Tasa de éxito")
fig.supxlabel("Episodios")

fig.suptitle(r"Recompensa normal de Gymnasium ($\gamma=1$)")

fig.suptitle(r"Recompensa modificada ($\gamma=0.95$), ")
plt.savefig("blackjack_sim_nr.pdf", backend='pgf')
plt.show()
plt.close()
```
